1. What is NumPy, and why is it widely used in Python?

-> NumPy (Numerical Python) is a powerful library in Python used for numerical computing. It was developed in 2005 by Travis Oliphant as an extension of the Numeric and Numarray libraries.


-> NumPy arrays (ndarrays) store data efficiently in a structured format. They support homogeneous data types, meaning all elements in a NumPy array must be of the same type.


-> Speed and Performance: NumPy is implemented in C, which makes its computations much faster compared to native Python lists. Since C is a low-level language, NumPy takes advantage of optimized algorithms and memory handling.


-> Memory Efficiency: NumPy arrays store data in a contiguous memory block, reducing overhead and improving access speed. This leads to faster computation compared to Python lists.


-> Vectorized Operations: NumPy supports vectorized operations, meaning mathematical operations can be performed on entire arrays at once instead of using loops, making it highly efficient for data manipulation.

```python
# Example 1

import numpy as np

l = [1, 2, 3, 4, 5]
arr = np.array(l)

type(arr)
```

2. How does broadcasting work in NumPy?

-> Broadcasting allows NumPy to perform operations on arrays of different shapes without creating extra copies of data.

-> When two arrays have different shapes, NumPy automatically expands the smaller array to match the shape of the larger one (if possible).

-> This helps in performing element-wise operations efficiently, without needing loops or manual resizing.

```python
#example 2

import numpy as np

A = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
B = np.array([1, 2, 3])  # Shape (1,3)

result = A + B
print(result)
```

4. Explain the use of the groupby() method in Pandas?

-> The groupby() method in Pandas is used to **group data** based on one or more columns and then **perform operations** like sum, mean, count, etc., on each group.

-> It is useful for **analyzing and summarizing** large datasets efficiently.

-> **How it works:**

a) **Splits the data** into groups based on a selected column.

b) **Applies a function** (like sum, mean, count) to each group.

c) **Combines** the results into a new DataFrame

```python
#Example 4

import pandas as pd

# Creating a sample DataFrame
data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Sales': [100, 200, 150, 250, 300]}

df = pd.DataFrame(data)

# Using groupby() to sum sales by category
grouped_df = df.groupby('Category')['Sales'].sum()
print(grouped_df)
```

6. What are the differences between NumPy arrays and Python lists?

-> Numpy arrays stores homogeneous data (all elements must be of the same type)	whereas Python lists stores heterogeneous data (elements can be of different types).

-> Numpy arrays are faster because it is implemented in C and optimized for numerical operations whereas Python lists are	slower because it is a general-purpose data structure.

-> Numpy arrays uses less memory as data is stored in a contiguous block whereas Python lists uses more memory due to additional overhead.

-> Numpy arrays cannot be changed in place (need to create a new array for modifications)	whereas Python lists can be changed in place (elements can be modified directly).

-> Numpy arrays are designed for mathematical and scientific computing	Python lists are more flexible but not optimized for numerical calculations

```python
#Example 6

import numpy as np

# NumPy array (homogeneous)
arr = np.array([1, 2, 3, 4])

# Python list (heterogeneous)
lst = [1, "hello", 3.5, True]

print(type(arr))
print(type(lst))
```

21) Why is NumPy's array slicing faster than Python's list slicing?

-> NumPy arrays are much faster than Python lists for slicing due to the following reasons:

   - Memory Efficiency
     - Contiguous Memory Storage
        - NumPy arrays are stored in a contiguous block of memory, unlike Python lists, which store references to objects.
        - This allows for fast retrieval and efficient slicing.

   - Optimized Implementation
     - NumPy Uses C for Computation
       - NumPy is built on C, a low-level language optimized for speed.
       - Python lists involve high-level dynamic type checking, which slows down slicing.

- No Type Checking Overhead
  - Homogeneous Data in NumPy
    - NumPy arrays store only one data type, eliminating the need for repeated type checking during slicing.
    - Python lists store heterogeneous data, requiring extra checks for each element.

- View-Based Slicing (No Copying)
  - Slicing Returns a View, Not a Copy
    - NumPy does not copy data when slicing—it just creates a "view" on the original array.
    - Python lists, however, create a new list copy when sliced, making it slower.

```python
#3. Create a scatter plot using Matplotlib.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Age': [25, 32, 40, 28, 35],
    'Salary': [50000, 70000, 80000, 55000, 60000],
    'Experience': [2, 5, 8, 3, 6]
}

df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```
